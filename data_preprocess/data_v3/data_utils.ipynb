```python
import pyodbc

query = """  

SELECT  px.fsym_id 
      , px.p_date 
      , px.exch_rate_sec_eur 
      , px.p_price_adj_split_spinoff_div 
      , px.mcap_local 
      ,sec.factset_industry 
      ,sec.factset_sector 
  FROM [ScreenFactory].[dbo].[03_DailyPx_Temp] px 
  left join [ScreenFactory].[dbo].[Securities_Fds] sec 
  on px.fsym_id = sec.fsym_id 
  WHERE px.p_date >= '1999-01-07' AND px.p_date <= '2019-12-26' 
"""
query_beta = """
SELECT  px.fsym_id
      , px.p_date
      , px.exch_rate_sec_eur
      , px.p_price_adj_split_spinoff_div
      , px.mcap_local
      ,sec.factset_industry
      ,sec.factset_sector
	  ,[04_Factor_Raw_Histo].[factor_value] as beta

  FROM [ScreenFactory].[dbo].[03_DailyPx_Temp] px

  left join [ScreenFactory].[dbo].[Securities_Fds] sec
  on px.fsym_id = sec.fsym_id

  left join [ScreenFactory].[dbo].[04_Factor_Raw_Histo] 
  on px.fsym_id = [04_Factor_Raw_Histo].fsym_id AND px.p_date = [04_Factor_Raw_Histo].p_date
  WHERE px.p_date >= '1998-01-01' AND px.p_date <= '2019-12-31' AND factor_raw_id = 255 
  order by px.p_date desc

"""



query_symbols_correspondence = """
        SELECT [sym_sedol_hist].[fsym_id]
              ,[bbg_ticker]
              ,[sedol]
              ,[fsym_security_id]
              ,[ce_sec_entity].factset_entity_id
              ,factset_ultimate_parent_entity_id
        FROM [fds].[sym_v1].[sym_sedol_hist]

        join [fds].[sym_v1].[sym_coverage]
        on [sym_coverage].fsym_id = [sym_sedol_hist].[fsym_id]

 

        join [fds].[sym_v1].[sym_bbg]
        on sym_bbg.fsym_id = sym_sedol_hist.fsym_id

 

        join [fds].[evt_v1].[ce_sec_entity]
        on [fsym_security_id] = [ce_sec_entity].fsym_id

 

        join [fds].[ent_v1].[ent_entity_structure]
        on [ce_sec_entity].factset_entity_id = [ent_entity_structure].factset_entity_id
    """
```

```python
df = df.rename(columns= 
               {'beta':'DES_BETA_ADJ_104W', 'mcap_local' : 'DES_MCAP_CCY', 'p_date':'DATE', 'fsym_id' : 'SYMBOL'}
               )
if 'DATE' in df.columns and df['DATE'].dtype == 'object':
    df['DATE'] = pd.to_datetime(df['DATE'])
    df = df.set_index(['SYMBOL','DATE']).sort_index()
else:
    assert "DATE" in df.index.names and df.index.get_level_values("DATE").dtype == 'datetime64[ns]'


# Calculate the USD adjusted price
usd_adjusted_price = df['p_price_adj_split_spinoff_div'] * df['exch_rate_sec_eur']

# Calculate the forward 5-day return
df['RET_5D'] = usd_adjusted_price.shift(-5) /usd_adjusted_price - 1
# df['RET_5D_dec'] = df.groupby('DATE')['RET_5D'].transform(lambda x:np.floor((x.rank(pct=True)-1e-7) * 10 ))
df['RET_5D_bin'] = df.groupby('DATE')['RET_5D'].transform(lambda x:np.floor((x.rank(pct=True)-1e-7) * 2))

#for 10D 
df['RET_10D'] = usd_adjusted_price.shift(-10) /usd_adjusted_price - 1
# df['RET_10D_dec'] = df.groupby('DATE')['RET_10D'].transform(lambda x:np.floor((x.rank(pct=True)-1e-7) * 10 ))
df['RET_10D_bin'] = df.groupby('DATE')['RET_10D'].transform(lambda x:np.floor((x.rank(pct=True)-1e-7) * 2))

df = df.rename(columns = { "factset_industry": "DES_FACTSET_INDUSTRY", "factset_sector": "DES_FACTSET_SECTOR"} )
df.to_parquet(os.path.join(data_folder, 'df_price.parquet'))
```

```python

import datetime, re
old_cols = ["mcap_local", "factset_industry", "factset_sector"]
new_cols =["DEs_MCAP_CCY", "DES_FACTSET_INDUSTRY", "DES_FACTSET_SECTOR"]
if any([col in df_price.columns for col in old_cols]):
    print(f"renaming columns {old_cols} to {new_cols}")
    df_price = df_price.rename(columns= dict(zip(old_cols, new_cols)))


#create SYMBOL index
print(f"creating SYMBOL index")
news_pkl = news_pkl.rename(columns = {"all_factset_entity_id": "SYMBOL", "publish_time": "DATE"})
news_pkl = news_pkl.explode("SYMBOL").dropna(subset = ['SYMBOL']) # converts list of symbols into separate rows
news_pkl = news_pkl[['DATE', 'SYMBOL', 'headline', 'story_body']]

# process DATE : convert 2012-06-06T05:40:39Z to 2012-06-06
print(f"processing DATE")
datecol = pd.to_datetime(news_pkl['DATE']).dt.tz_localize(None).dt.strftime('%Y-%m-%d')
news_pkl['DATE'] = pd.to_datetime(datecol)

#set the index to SYMBOL, DATE
print(f"setting index to SYMBOL, DATE")
news_pkl = news_pkl.set_index(['SYMBOL', 'DATE']).sort_index()

print(f"shifting dates by 1 day")
news_pkl['DATE2'] =( news_pkl.reset_index(level = 1)['DATE'] + datetime.timedelta(days=1)).values
news_pkl = news_pkl.reset_index(level = 1, drop = True).rename(columns = {"DATE2": "DATE"}).reset_index(drop=False).set_index(['SYMBOL', 'DATE']).sort_index()
```

```python

print('removing "StreetAccount" reports ')
pattern = r'.*Street\s*Account.*'
def filter_function(x:str):
    if not x.__class__ == str:
        return False
    return re.search(pattern, x) is not None
news_pkl['hasSA'] = news_pkl['headline'].apply(filter_function)
print("num 'StreetAccount' rows : ", news_pkl.hasSA.sum() / news_pkl.shape[0], "out of ", news_pkl.shape[0])
news_pkl = news_pkl.loc[~news_pkl.hasSA, ["headline", "story_body"]]
```
